```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
from google.colab import drive

drive.mount('/content/drive')
```

```python
path = "/content/drive/MyDrive/csv Files/company_sales_data.csv"
```

```python
df=pd.read_csv(path)  
```

Plot Jointplot and visualise the data.

```python
sns.jointplot(x='month_number',y='total_profit',kind='scatter',data=df)
```

Plot Distplot , make kde to false and make bins to 30.

```python
sns.distplot(df['total_profit'], kde=False, bins=30, color='red')
```

Plot Scatterplot , try adding hue in it.

```python
sns.scatterplot(data=df,x='month_number',y='total_profit',hue='total_units')
```

Plot swarmplot and change color palette to ‘VIBGYOR’.

```python
colors = [	"#9400D3",	"#9400D3",	"#0000FF",	"#00FF00",	"#FFFF00",	"#FF7F00",	'#FF0000']
sns.set_palette(sns.color_palette(colors))
```

```python
sns.swarmplot(data=df,x='month_number',y='total_profit')
```

Plot countplot and use hue parameter in it.

```python
sns.countplot(x='shampoo',hue='total_profit',data=df)
```

Correlate the data and plot Heatmap on it and change its color scale to any you
like.

```python
sns.heatmap(df)
```

Plot a facetgrid and use hue in it.

```python
g= sns.FacetGrid(data=df,col='total_profit',hue='total_units')
g.map_dataframe(sns.scatterplot, x='month_number',y='shampoo')
```

Plot a pairplot

```python
sns.pairplot(df)
```
